```python
import cv2
import glob
import os
import matplotlib.pyplot as plt
from matplotlib import patches
from PIL import Image
import numpy as np
```

```python
faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
path = "faces/train/angry/*.jpg"
```

```python
count = 0
images = []
```

```python
for faces in glob.glob(path, recursive=True):
    images.append(faces)
    count += 1
    if count > 10:
        break
```

```python
images[0]
```

```python
img = cv2.imread('people.jpg')
# Convert the image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray_image.shape
```

```python
plt.figure(figsize=(12,8))
```

```python
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
faces = faceCascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=12, flags=cv2.CASCADE_SCALE_IMAGE)
```

```python
print("Founded face is {} faces which are {}".format(len(faces), faces))
```

```python
num_of_people = len(faces)
for (x1, y1, w1, h1) in faces:
    cv2.rectangle(gray_image, (x1, y1), (x1+w1, y1+h1), (255,0,0), 2)
    face_image = img[y1:y1+h1, x1:x1+w1]
    cv2.imwrite(f'crop{num_of_people}.jpg', face_image)
    num_of_people -= 1
```

```python
plt.figure(figsize=(12,8))
plt.imshow(gray_image, cmap='gray')
plt.show()
```
